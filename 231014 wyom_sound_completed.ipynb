```python
import requests
import urllib3
from bs4 import BeautifulSoup


```

```python
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# URL of the webpage
# url = "https://weather.uwyo.edu/cgi-bin/sounding?region=seasia&TYPE=TEXT%3ALIST&YEAR=2023&MONTH=10&FROM=0100&TO=3112&STNM=47122"

url = "https://weather.uwyo.edu/cgi-bin/sounding"

# Define the query parameters
params = {
    'region': 'seasia',
    'TYPE': 'TEXT:LIST',
    'YEAR': '2022',
    'MONTH': '01',
    'FROM': '0100',
    'TO': '3112',
    'STNM': '47122'
}

response = requests.get(url, params=params, verify=False)

soup = BeautifulSoup(response.content, "html.parser") # 보기좋게 만들것
```

```python
html_list = soup.find_all("html")
body_list = soup.find_all("body")
h2_list = soup.find_all("h2") #.find_all("pre")
pre_list = soup.find_all("pre") #.find_all("pre")
p_list = soup.find("p").decompose()

soup.fi

html_list_1 = []
body_list_1 = []
h2_list_1 = []
pre_list_2 = []

for html in html_list:
    html_list_1.append(html.text)

for body in body_list:
    body_list_1.append(body.text)
    

for h2 in h2_list:
    h2_list_1.append(h2.text)


for pre in pre_list:
    pre_list_2.append(pre.text)

del p_list

print(body_list_1)

```
